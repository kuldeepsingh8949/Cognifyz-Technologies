# *Analyze the relationship between the type ofcuisine and the restaurant's rating.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\kuldeep singh\OneDrive\Desktop\Cognifyz-Technologies\dataset.csv')
df.head()
```

```python
df[['Cuisines', 'Aggregate rating']].head()
```

```python
df_exploded = df.dropna(subset=['Cuisines']).copy()
df_exploded.head()
```

```python
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.split(', ')
df_exploded = df_exploded.explode('Cuisines')
```

```python
cuisine_rating = df_exploded.groupby('Cuisines')['Aggregate rating'].agg(['mean', 'count']).reset_index()
cuisine_rating
```

```python
cuisine_rating.columns = ['Cuisine', 'Average Rating', 'Count']
cuisine_rating_sorted = cuisine_rating.sort_values(by='Average Rating', ascending=False)
```

```python
top_cuisines = cuisine_rating_sorted[cuisine_rating_sorted['Count'] >= 20].head(10)
top_cuisines
```

# *Identify the most popular cuisines among customers based on the number of votes.*

```python
popular_cuisines = df_exploded.groupby('Cuisines')['Votes'].sum().reset_index()
popular_cuisines
```

```python
popular_cuisines_sorted = popular_cuisines.sort_values(by='Total Votes', ascending=False)
popular_cuisines_sorted
```

```python
top_voted_cuisines = popular_cuisines_sorted.head(20)
top_voted_cuisines
```

# *Determine if there are any specific cuisines that tend to receive higher rating*

```python
df_exploded = df.dropna(subset=['Cuisines']).copy()
df_exploded.head()
```

```python
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.split(', ')
df_exploded = df_exploded.explode('Cuisines')
```

```python
cuisine_rating = df_exploded.groupby('Cuisines')['Aggregate rating'].agg(['mean', 'count']).reset_index()
cuisine_rating
```

```python
cuisine_rating.columns = ['Cuisine', 'Average Rating', 'Restaurant Count']
```

```python
filtered_cuisine_rating = cuisine_rating[cuisine_rating['Restaurant Count'] >= 20]
filtered_cuisine_rating.head()
```

```python
top_rated_cuisines = filtered_cuisine_rating.sort_values(by='Average Rating', ascending=False)
top_rated_cuisines.head()
```
