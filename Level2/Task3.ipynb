# *Extract additional features from the existingcolumns, such as the length of the restaurantname or address*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\kuldeep singh\OneDrive\Desktop\Cognifyz-Technologies\dataset.csv')
df.head()
```

```python
df['Restaurant_length'] = df['Restaurant Name'].astype(str).apply(len)
df.head()
```

```python
df['address_length'] = df['Address'].astype(str).apply(len)
df.head()
```

```python
df['name_word'] = df['Restaurant Name'].astype(str).apply(lambda x: len(x.split()))
df.head()
```

# *Create new features like "Has Table Booking"or "Has Online Delivery" by encodingcategorical variables.*

```python
df['Has_Table_Booking'] = df['Has Table booking'].str.lower().map({'yes': 1, 'no': 0})
df.head()
```

```python
df['Has_Online_Delivery'] = df['Has Online delivery'].str.lower().map({'yes': 1, 'no': 0})
df.head()
```
