*Perform data type conversion if necessary.Analyze the distribution of the target variable("Aggregate rating") and identify any classimbalances*

```python
df.dtypes
```

```python
# Convert Object Columns to Category 

categorical_columns = [
    'City', 'Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery',
    'Is delivering now', 'Switch to order menu', 'Rating color', 'Rating text'
]
df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
df.dtypes
```

```python
# Analyze Distribution of 'Aggregate rating'
agg = df['Aggregate rating'].value_counts().sort_index()
agg 
```

```python
agg_percentages = df['Aggregate rating'].value_counts().sort_index()*100
agg_percentages
```

```python
# Identify Class Imbalance
rating_summary = pd.DataFrame({
    'Rating': agg.index,
    'Count': agg.values,
    'Percentage': agg_percentages.values.round(2)
}).reset_index(drop=True)
rating_summary
```

```python
# Plot chart of 'Aggregate rating'
plt.figure(figsize=(10, 6))
sns.countplot(x='Aggregate rating', data=df, order=sorted(df['Aggregate rating'].unique()))
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
