#*Explore the distribution of categorical variables like "Country Code," "City," and"Cuisines.*

```python
categorial = df.select_dtypes(['object'])
categorial
```

```python
plt.figure(figsize=(10, 4))
df['Country Code'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Country Codes')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
df['City'].value_counts().head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 4))
df['Cuisines'].value_counts().head(10).plot(kind='bar', color='salmon')
plt.title('Top 10 Cuisines Combinations')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.show()
```

#*Identify the top cuisines and cities with thehighest number of restaurants.*

```python
from collections import Counter
```

```python
cuisine_series = df['Cuisines'].dropna().str.split(', ')
cuisine_list = [cuisine for sublist in cuisine_series for cuisine in sublist]
cuisine_counts = Counter(cuisine_list)
top_cuisines = cuisine_counts.most_common(10)
top_cities = df['City'].value_counts().head(10)
```
